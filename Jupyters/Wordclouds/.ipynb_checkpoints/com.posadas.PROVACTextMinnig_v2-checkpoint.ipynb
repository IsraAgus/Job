```python
###
# Concatenacion de  los dos dataframes
###
frames = [comment_1,comment_2]
df = pd.concat(frames)
print(df.shape)
df.head()
```

```python
###
# Limpieza de los comentarios
# Se separa el comentario del socio que llamó y se limpia con regex
# a su vez se devuelven dos listas, la del usuario que llamó y el c
# omentario
###
l_k = []
l_v = []
dicc = {}
count1 = 0
count2 = 0
for item in df.description.values.tolist():
    if re.search('^[\w]+:',item) != None: #Solo letras del alfabeto incluidas letras francesas y español
        count1 += 1
        c1,c2 = re.search('^[\w]+:',item).span()
        l_k.append(item[0:c2-1].lower())
        l_v.append(
            re.sub(
                '[^\w0-9 ]+', #Permitimos el paso de letas y numeros
                " ",
                item[c2+1:-1].strip().lower()
            )
        )
    else:
        count2 += 1
        l_k.append("no_user")
        l_v.append(item)
```

<p>En el proceso de la limpieza de la información, RegEx es una herrmienta muy buena para poder hacer la limpieza de los datos. 
    <ul>
        <li>'\b' permite indicar un borde de palabra, es decir, buscar la palabra sea una cadena independiente y no una subcadena de caracteres.
        <li> [] los corchetes indican un set de caracteres
        <li> ^ el gorro indica el inicio de una cadena
        <li> '\w' indica que solo pasarán caracteres alfabeticos incluyendo alfabeto francés y español
        <li> 0-9 indica que pasarán caracteres numericos
    </ul>
</p>

```python
###
# De la lista de comentarios limpiamos los cadenas dis, dispo
# disp y sol
###
l_aux = []
for item in l_v:
    if re.search(r'\bdis\b',item) != None: #Si se encontro la cadena
        pos1,pos2 = re.search(r'\bdis\b',item).span() #Encuentra la cadena y dame las posiciones donde se encuentra (inicio,final)
        str1 = list(item)[0:pos1] # guarda la primera parte de la oración antes de la coincidencia
        str2 = list(item)[pos2:-1] # guarda la segunda parte de la oración despúes de la coincidencia
        if re.search(r'\bsol\b',item) != None: #Si encuentras dentro del texto la cadena sol
            pos1,pos2 = re.search(r'\bsol\b',item).span() #Dame las pisiciones inicio y fin
            str1 = list(item)[0:pos1] #guarda la primera parte d ela oración antes d ela coincidencia
            str2 = list(item)[pos2:-1] #guarda la segunda parte d ela oración despúes de la coincidencia
            l_aux.append("".join(str1+list('solicita')+str2)) #agregala a la lista, la cadena limpia sustityendo sol por solicita
        else: #si no se encontro sol
            l_aux.append("".join(str1+list('disponibilidad')+str2)) #solo guarda la cadena limpia
    elif re.search(r'\bdisp\b',item) != None:
        pos1,pos2 = re.search(r'\bdisp\b',item).span()
        str1 = list(item)[0:pos1]
        str2 = list(item)[pos2:-1]
        if re.search(r'\bsol\b',item) != None:
            pos1,pos2 = re.search(r'\bsol\b',item).span()
            str1 = list(item)[0:pos1]
            str2 = list(item)[pos2:-1]
            l_aux.append("".join(str1+list('solicita')+str2))
        else:
            l_aux.append("".join(str1+list('disponibilidad')+str2))
    elif re.search(r'\bdispo\b',item) != None:
        pos1,pos2 = re.search(r'\bdispo\b',item).span()
        str1 = list(item)[0:pos1]
        str2 = list(item)[pos2:-1]
        if re.search(r'\bsol\b',item) != None:
            pos1,pos2 = re.search(r'\bsol\b',item).span()
            str1 = list(item)[0:pos1]
            str2 = list(item)[pos2:-1]
            l_aux.append("".join(str1+list('solicita')+str2))
        else:
            l_aux.append("".join(str1+list('disponibilidad')+str2))
    else:
        l_aux.append(item)
l_v = l_aux
l_aux = []
```

```python
###
# Repitiendo el proceso anterior para forzar la limpieza de los datos
###
l_aux = []
for item in l_v:
    if re.search(r'\bdis\b',item) != None:
        pos1,pos2 = re.search(r'\bdis\b',item).span()
        str1 = list(item)[0:pos1]
        str2 = list(item)[pos2:-1]
        if re.search(r'\bsol\b',item) != None:
            pos1,pos2 = re.search(r'\bsol\b',item).span()
            str1 = list(item)[0:pos1]
            str2 = list(item)[pos2:-1]
            l_aux.append("".join(str1+list('solicita')+str2))
        else:
            l_aux.append("".join(str1+list('disponibilidad')+str2))
    elif re.search(r'\bdisp\b',item) != None:
        pos1,pos2 = re.search(r'\bdisp\b',item).span()
        str1 = list(item)[0:pos1]
        str2 = list(item)[pos2:-1]
        if re.search(r'\bsol\b',item) != None:
            pos1,pos2 = re.search(r'\bsol\b',item).span()
            str1 = list(item)[0:pos1]
            str2 = list(item)[pos2:-1]
            l_aux.append("".join(str1+list('solicita')+str2))
        else:
            l_aux.append("".join(str1+list('disponibilidad')+str2))
    elif re.search(r'\bdispo\b',item) != None:
        pos1,pos2 = re.search(r'\bdispo\b',item).span()
        str1 = list(item)[0:pos1]
        str2 = list(item)[pos2:-1]
        if re.search(r'\bsol\b',item) != None:
            pos1,pos2 = re.search(r'\bsol\b',item).span()
            str1 = list(item)[0:pos1]
            str2 = list(item)[pos2:-1]
            l_aux.append("".join(str1+list('solicita')+str2))
        else:
            l_aux.append("".join(str1+list('disponibilidad')+str2))
    else:
        l_aux.append(item)
l_v = l_aux
l_aux = []
```

```python
###
# Se crean dos columnas nuevas con las listas de usuario y comentario
###
df['usuarios'],df['comentario_limpio'] = l_k,l_v
```

<div class="alert alert-info" role="alert">
    <h3>Crear los n-grams</h3>
    <hr>
    <p>Básicamente es otro punto para poder entender que información esta manipulandoce</p>
</div>

<p>
    Por medio del módulo de NLTK utilizamos la clase de ngrmas, la cual a través de un texto continuo e indicando el númeor de ngrams nos devolverá un diccionario con las frecuencias de las palabras.
    Luego, para la presentación de la información se utiliza una creación de un diccionario y un DF, como se ve en el siguiente algoritmo:
    <ul>
        <li> Creación del diccionario para crear el dataframe (1)
        <li> Creación de una nueva columna para el n el contador de ocurrencias (2)
        <li> Se instancia el dataframe con el diccionario anterior (3)
        <li> Vamos creando el mismo diccionario anterior pero en ejecución por cada palabra, ocurrencia se crea el diccionario y se manda a agregar al data frame (4)
        <li> Creamos una nueva columna que se concatenará las palabras
        <li> Reordenamos el datframe por nombre de columnas 
        <li> Reacomodamos las columnas
        <li> Guardamos el archivo
    </ul>
</p>

```python
esBigrams = ngrams(''.join(df.comentario_limpio.values.tolist()).split(), 3) #Este igual es un input, el tres
esBigramFreq = collections.Counter(esBigrams)
lista_ngrams = esBigramFreq.most_common(50)
size_n = 2 #Este puede ser un input, de hecho deberia de serlo
dicc = {i:j for i,j in zip(['word'+str(i+1) for i in range(size_n)],[[] for i in range(size_n)])} # 1
dicc['n'] = [] # 2
df_ngram = pd.DataFrame(dicc) # 3
for item in lista_ngrams: #4
    dicc = {i:j for i,j in zip(['word'+str(i+1) for i in range(len(item[0]))],[v for v in item[0]])} #4
    dicc['n'] = item[1] #4
    df_ngram = df_ngram.append(dicc,ignore_index=True,sort=True) #4
df_ngram['concatenacion'] = df_ngram['word1'] + '-' + df_ngram['word2'] + '-' + df_ngram['word3'] #5
df_ngram = df_ngram.reindex(sorted(df_ngram.columns,reverse=False), axis=1) #6
df_ngram = df_ngram[['word1','word2','word3','concatenacion','n']] #7


df_ngram.to_csv('3-grams_comma.csv',index=False,encoding='utf-8-sig',quoting=csv.QUOTE_ALL) #7
df_ngram.to_csv('3-grams_pipe.csv',index=False,sep='|',encoding='utf-8-sig',quoting=csv.QUOTE_ALL) #7

df_ngram
```

<div class="alert alert-warning" role="alert">
    <h3>Algoritmo fecha de cambio de estatus de socio</h3>
    <p>Este algoritmo tiene la intención de poder generar un DF que contenga las fechas de cuando el socio llamo en la bitacora y cuando el tuvo un cambio en el estado de su membresia. 
       La intención es poder identificar el rango de fecha para poder segmentar comentarios y entender más a detalle el cambio tanto particular como general de socios que pasaron de activo a holding por ejemplo o en su defecto cualquier cambio de estado.
    El algoritmo se explica a continuación:
    Básicamente el algoritmo necesita generar las n columnas en tiempo de ejecución, es decir, se vana segmentar los socios del df creando una lista unica de socios desduplicados que al momento de extraer su información, creemos un df con la primer fecha registrada en una llamada del socio de esta bitacora y su estatus, luego iterar ese subconjunto de datos hasta encontrar un cambio en su estatus, guardando la nueva fecha y el nuevo estatos, así sucesivamente generando un historico de cambios, obviamente el resultado debería de ser un DF como una matriz dispersa, en la cual simularemos que las transacciones son los cambios de estado del socio y sus fechas. 
    Como bien puede ser el caso en el que los socios solo tengan un cambio en todo su historico ya sea por ser nuevos o otras razones, también puede haber socios que tengan un estatus activo, luego holding, luego cancelado, luego activo, etc. 
    Pasos: 
        <li> Creacion del DF (1)
        <li> Primer ciclo por cada miembro en la lista de miembros unicos: (2)
        <li>     seleccionamos de los datos originales todo el historico del socio y ordenamos los datos por la fecha, de la primer fecha hasta la más actual (3)
        <li>     En el dataframe registramos su primer fecha y estatus, siendo muy obvio que gracias al reordenamiento el primer registro sea el primer estatus y fecha (4)
        <li>     En el segundo ciclo, por cada estado y fecha en las listas de estados y fechas,  (5)
        <li>           Si se registro un cambio en el estatus que sea diferente al primer estatus, cuara las nuevas lomunas llamadas socio, estatus-n, fecha-n  y agregalas al DF en caso contrario, no hagas nada. (6)
        <li>           Por último, crea una nueva columna con el ultimo estado registrado en DF llamada z_last_status y z_last_date (7)
    </p>
</div>
