```python
profiles = pd.read_excel("Profile.xlsx")
profiles[['http','_','url','usr']] = profiles.ID.str.split('/',expand=True)
profiles
```

```python
usrs=list(profiles.usr)
```

```python
data = api.user_timeline(id="JavBarr")
```

```python
comments = {'usuario':[],
            'comentarios':[]}
df_coments = pd.DataFrame(comments)
for i in range(40):    
    try:
        data = api.user_timeline(id=usrs[i])
        for element in data:
            dicc = {'usuario':usrs[i],
                    'comentarios':element['text']}
            #print(dicc)
            df_coments = df_coments.append(dicc,ignore_index = True)
            #print(df_coments)
    except:
        dicc = {'usuario':usrs[i],
                'comentarios':"Not authorized."}
        #print(dicc)
        df_coments = df_coments.append(dicc,ignore_index = True)
        #print(df_coments)
```

```python
df_coments.to_csv("comentarios_pipe.csv",sep='|',quoting=csv.QUOTE_ALL,index=False,encoding="utf-8-sig")
df_coments.to_csv("comentarios_comma.csv",quoting=csv.QUOTE_ALL,index=False,encoding="utf-8-sig")
```
