```python
def _connect_hyatt():
    """Método utilizado para instanciar solo el driver
       Parametros: <None>
       Retorno: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado"""
    #url = "https://www.resortsbyhyatt.com/ziva-cancun?utm_source=Hyatt&utm_medium=link+off&utm_term=Ziva+Cancun&utm_campaign=Hyatt+Link+Off+PP"
    chromeOptions = webdriver.ChromeOptions()
    chromeOptions.add_argument("--incognito")
    driver = webdriver.Chrome("chromedriver.exe",options=chromeOptions)
    #driver.get(url)
    _wait(5)
    return driver

def _update_page(driver,date1,date2):
    """Método utilizado para actualizar la página web funciona igual que en Iberostar.
       Parámetros: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado
       Retorno: <None>"""
    url = "https://be.synxis.com/?adult=2&arrive="+str(date1)+"&chain=17001&child=0&configcode=PlayaWebsite&coupon=&currency=USD&depart="+str(date2)+"&gaclientid=538837421.1578407401&hotel=1460&level=hotel&local=en-US&locale=en-US&rate=WEBSPECIAL&rooms=1&sbe_sid=5Suk8-9Az5hOXwqw9_eHwokO&themecode=PlayaWebsite"
    driver.get(url)
    _wait(5)
    return None

def _click_show_rooms(driver):
    """Método utilizado para visualizar los cuartos y los precios de las habitaciones, hay que desplegar donde diga show rooms
       Párametros: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado
       Retorno: <None>"""
    str_show = 'div[class=thumb-cards_button]'
    try:
        buttons = driver.find_elements_by_css_selector(str_show)
        #buttons.pop(0)
        #buttons.pop(0)
        for element in buttons:
            element.click()
    except:
        print("Botones show rooms no encontrados")
    return None

def _get_prices_and_rooms(driver):
    """Future doc"""
    str_prices = 'ins[class="thumb-cards_price"]'
    str_rooms = 'div[class="thumb-cards_cardHeader"]'
    lista_rooms = []
    lista_prices = []
    try:
        #print("I'm here")
        rooms = driver.find_elements_by_css_selector(str_rooms)
        #print(len(rooms))
        for element in rooms:
            name = element.text.split('\n')
            lista_rooms.append(name[0])
            #print("Rooms: ", element.text)
    except:
        print("Elemento titles no encontrado")
    try:
        prices = driver.find_elements_by_css_selector(str_prices)
        for element in prices:
            #print("Prices: ", element.text)
            #price = element.text
            lista_prices.append(element.text.replace('$',''))
    except:
        print("Elemento prices no encontrado")
    return lista_rooms,lista_prices

def _firsbutton_hyatt(driver):
    """Método utilizado para hacer click en el boton del Hyatt
       Parámetros: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado
       Retorno: <None>"""
    str_button = 'button[class="btn btn-sm bg-ziva-pink"]'
    try:
        _wait(2)
        button = driver.find_element_by_css_selector(str_button)
        button[1].click()
        _wait(3)
    except:
        print("Elemento botón no encontrado")
    return None

def _choose_dates(driver,datein,dateout):
    """Método utilizado para seleccionar las fechas deseadas y hacer click en el envio
       Párametros: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado
                   <datein> tipo: datetime --> La fecha de llegada
                   <dateout> tipo: datetime --> La fecha de salida
       Retorno: <None>"""
    str_calendar_in = 'input[name="checkinDate"]'
    str_datein = 'td[data-iso-value="'+str(datein)+'"]'
    str_dateout = 'td[data-iso-value="'+str(dateout)+'"]'
    str_send = 'button[class="quickbookSearchFormButton button-shop button-block"]'
    try:
        calendar = driver.find_element_by_css_selector(str_calendar_in)
        calendar.click()
        _wait(1)
        try:
            datein = driver.find_element_by_css_selector(str_datein)
            datein.click()
            _wait(1)
            try:
                dateout = driver.find_element_by_css_selector(str_dateout)
                dateout.click()
                _wait(1)
                try:
                    button = driver.find_element_by_css_selector(str_send)
                    print(button.text)
                    button.click()
                    _wait(1)
                except:
                    print("Elemento button no encontrado")
            except:
                print("Elemento tab dateout no encontrado")
        except:
            print("Elemento tab datein no encontrado")
    except:
        print("Elemento calendarin no encontrado")
    return None

def _back_page(driver):
    """Método utilizado para regresar un paso atras
       Parámetros: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado
       Retorno: <None>"""
    driver.execute_script("window.history.go(-1)")
    _wait(3)
    return None

def _wait(num):
    """Método utilizado para realizar una espera implicita
       Parámetros: <num> tipo: int --> Requiere de un número para indicar una espera
       Retorno: <None>"""
    time.sleep(num)

def _quit(driver):
    """Método utilizado para cerrar el driver
       Parámetros: <selenium.webdriver> tipo: selenium.webdriver --> Es el bot ya levantado y configurado
       Retorno: <None>"""
    driver.quit()
    return None
```

```python
###
# Obtención del precio del dólar
###
c = CurrencyRates()
mxn = c.get_rate('USD', 'MXN')
mxn
```

```python
num_days = int(input("Ingresa el número de días a consultar: "))
dicc_hyatt = {"Hotel":[],
             "Fecha_consulta":[],
             "Fecha_reserva_checkin":[],
             "Fecha_reserva_checkout":[],
             "Nombre_habitacion":[],
             "Precio_habitacion_MXN":[],
             "Precio_habitacion_USD":[],
             "Precio_dolar_al_dia":[]}
hyatt_data = pd.DataFrame(dicc_hyatt)
consult_Date = date.today()
today = datetime.date.today()
tomorrow = today + datetime.timedelta(days=1)
driver = _connect_hyatt()
_update_page(driver,today,tomorrow)
while num_days > 0:
    _click_show_rooms(driver)
    habitaciones,precios=_get_prices_and_rooms(driver)
    for habitacion,precio in itertools.zip_longest(habitaciones,precios):
        print(habitacion,precio)
        if precio == None:
            precio = '0'
        else: 
            if "," in precio:
                precio = precio.replace(",","")
        dicc = {"Hotel":"Hyatt Ziva Cancún",
                 "Fecha_consulta":str(consult_Date),
                 "Fecha_reserva_checkin":str(today),
                 "Fecha_reserva_checkout":str(tomorrow),
                 "Nombre_habitacion":habitacion,
                 "Precio_habitacion_MXN":float(precio)*mxn,
                 "Precio_habitacion_USD":precio,
                 "Precio_dolar_al_dia":mxn}
        hyatt_data = hyatt_data.append(dicc,ignore_index = True) 
    today = tomorrow
    tomorrow = today + datetime.timedelta(days=1)
    _update_page(driver,today,tomorrow)
    num_days -= 1
_quit(driver)
```

```python
###
# Limpiando las habitaciones de nuestro interes
###
hyatt_data = hyatt_data.loc[(hyatt_data['Nombre_habitacion']=='OCEAN FRONT DOUBLE') |\
                     (hyatt_data['Nombre_habitacion']=='OCEAN VIEW DOUBLE') |\
                     (hyatt_data['Nombre_habitacion']=='CLUB OCEAN FRONT DOUBLE'),:]
```
